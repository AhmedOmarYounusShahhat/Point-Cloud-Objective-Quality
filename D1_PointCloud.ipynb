```python
def find_closest(x_array, y_array):
    Sum=0
    i=0
    k=[]
    remaining_x_array = x_array[:]
    for y in y_array:
        differences = []
        for x in remaining_x_array:
            differences.append(((y - x)**2).sum())
        # min_index_remaining is the index position of the closest x value
        # to the given y in remaining_x_array
        min_index_remaining = differences.index(min(differences))
        Sum += min(differences)
        k.append(min_index_remaining)
    return Sum
```

```python
def find_closest_c(x_arrayx, y_arrayx):
    Sum=0
    k=[]
    R=0
    G=0
    B=0
    i=0
    remaining_x_array = x_arrayx[:,:3]
    x_c = x_arrayx[:,3:]
    y_array= y_arrayx[:,:3]
    y_c = y_arrayx[:,3:]
    for y in y_array:
        differences = []
        
        for x in remaining_x_array:
            differences.append(((y - x)**2).sum())
        # min_index_remaining is the index position of the closest x value
        # to the given y in remaining_x_array
        min_index_remaining = differences.index(min(differences))
        R+= ((y_c[:,:1][i]-x_c[:,:1][min_index_remaining])**2)[0]
        G+= ((y_c[:,1:2][i]-x_c[:,1:2][min_index_remaining])**2)[0]
        B+= ((y_c[:,2:3][i]-x_c[:,2:3][min_index_remaining])**2)[0]
        Sum += min(differences)
        i+=1
        k.append(min_index_remaining)
    return Sum,R,G,B
```
